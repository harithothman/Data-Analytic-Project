##Forecasting Horizon

```python
def create_features(prod_malaysia):

    prod_malaysia = prod_malaysia.copy()
    prod_malaysia['year'] = prod_malaysia.index.year
    return prod_malaysia

prod_malaysia = create_features(prod_malaysia)
```

##Lag Features

```python
def add_lags(prod_malaysia):
    target_map = prod_malaysia['Emissions from production'].to_dict()
    prod_malaysia['lag1'] = (prod_malaysia.index - pd.Timedelta('364 days')).map(target_map)
    prod_malaysia['lag2'] = (prod_malaysia.index - pd.Timedelta('728 days')).map(target_map)
    prod_malaysia['lag3'] = (prod_malaysia.index - pd.Timedelta('1092 days')).map(target_map)
    return prod_malaysia
```

```python
prod_malaysia = add_lags(prod_malaysia)
```

```python
prod_malaysia
```

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
tss = TimeSeriesSplit(n_splits=3)
prod_malaysia = prod_malaysia.sort_index()


fold = 0
preds = []
scores = []
for train_idx, val_idx in tss.split(prod_malaysia):
    train = prod_malaysia.iloc[train_idx]
    test = prod_malaysia.iloc[val_idx]

    train = create_features(train)
    test = create_features(test)

    FEATURES = ['year',
                'lag1','lag2','lag3']
    TARGET = 'Emissions from production'

    X_train = train[FEATURES]
    y_train = train[TARGET]

    X_test = test[FEATURES]
    y_test = test[TARGET]

    reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',
                           n_estimators=1000,
                           early_stopping_rounds=50,
                           objective='reg:linear',
                           max_depth=3,
                           learning_rate=0.01)
    reg.fit(X_train, y_train,
            eval_set=[(X_train, y_train), (X_test, y_test)],
            verbose=100)

    y_pred = reg.predict(X_test)
    preds.append(y_pred)
    score = np.sqrt(mean_squared_error(y_test, y_pred))
    scores.append(score)
```
